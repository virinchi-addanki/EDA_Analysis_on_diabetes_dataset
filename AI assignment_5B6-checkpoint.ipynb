```python
import matplotlib.pyplot as plt
import seaborn as sns

fig1, ax1 = plt.subplots(1, 2, figsize=(20, 7))

sns.histplot(data=df, x="Pregnancies", kde=True, ax=ax1[0], color='magenta')
sns.boxplot(data=df, x="Pregnancies", ax=ax1[1], color='aqua')

plt.show()
```

```python
fig3, ax3 = plt.subplots(1, 2, figsize=(20, 7))
sns.histplot(data=df, x="Glucose", kde=True, ax=ax3[0], color = 'magenta')
sns.barplot(data=df, x="Pregnancies", ax=ax3[1], color='lime')
plt.show()
```

```python
print("Median of Blood Pressure: ", df["BloodPressure"].median())
print("Maximum of Blood Pressure: ", df["BloodPressure"].max())
print("Mean of Pressure: ", df["BloodPressure"].mean())
```

```python
grouped_data = df.groupby('Age')['Glucose'].mean()
print(grouped_data)
```

# Distribution of Mean Glucose levels by Age Group

```python
mean_glucose_values = grouped_data.values

# Plot the histogram using the extracted mean glucose values.
plt.hist(mean_glucose_values, bins=11)  # You can adjust the number of bins as per your preference.
plt.xlabel('Mean Glucose')
plt.ylabel('Frequency')
plt.title('Distribution of Mean Glucose levels by Age Group')
plt.show()
```

```python
grouped_data_df = grouped_data.reset_index()

# Sort the DataFrame by 'Age' in ascending order for a better line plot.
grouped_data_df.sort_values(by='Age', inplace=True)

# Extract the 'Age' and 'Glucose' columns for plotting.
ages = grouped_data_df['Age']
mean_glucose_values = grouped_data_df['Glucose']

# Plotting the line plot.
plt.plot(ages, mean_glucose_values, marker='o', linestyle='-', color='b')
plt.xlabel('Age')
plt.ylabel('Mean Glucose')
plt.title('Mean Glucose levels by Age Group')
plt.grid(True)
plt.show()
```

```python
plt.bar(ages, mean_glucose_values)
plt.xlabel('Age')
plt.ylabel('Mean Glucose')
plt.title('Mean Glucose levels by Age Group')
plt.grid(axis='y')
plt.show()
```
